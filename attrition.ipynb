## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
column_names = ['Age', 'BusinessTravel', 'YearsAtCompany', 'DistanceFromHome', 'Education', 'EducationField', 'EnvironmentSatisfaction', 'JobSatisfaction', 'HourlyRate', 'JobInvolvement']


# Create X_df using your selected columns
X_df = attrition_df[column_names]
X_df.head()

# Show the data types for X_df
X_df.info()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
#encoder = OneHotEncoder(sparse=False)
#encoder.fit_transform(X_train)
X_train_bt = pd.get_dummies(X_train["BusinessTravel"])
X_test_bt = pd.get_dummies(X_test["BusinessTravel"])

X_train_bt.head()
```

```python
# Convert your X data to numeric data types however you see fit
X_train_edu = pd.get_dummies(X_train["EducationField"])
X_test_edu = pd.get_dummies(X_test["EducationField"])

X_train_edu.head()
```

```python
#displaying data to confirm encoding
X_train_encoded = pd.concat([X_train, X_train_bt, X_train_edu], axis=1)
X_test_encoded = pd.concat([X_test, X_test_bt, X_test_edu], axis=1)

X_train_encoded.head()
```

```python
#Droping original non numeric features
X_train_encoded = X_train_encoded.drop(columns=["BusinessTravel", "EducationField"])
X_test_encoded = X_test_encoded.drop(columns=["BusinessTravel", "EducationField"])

X_train_encoded.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train_encoded)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
# Create a OneHotEncoder for the Department column
encoder_department = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
encoder_department.fit(y_train[["Department"]])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_department = encoder_department.transform(y_train[["Department"]])
y_test_encoded_department = encoder_department.transform(y_test[["Department"]])

y_train_encoded_department
```

```python
# Create a OneHotEncoder for the Attrition column
encoder_attrition = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
encoder_attrition.fit(y_train[["Attrition"]])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_attrition = encoder_attrition.transform(y_train[["Attrition"]])
y_test_encoded_attrition = encoder_attrition.transform(y_test[["Attrition"]])

y_train_encoded_attrition
```

## Create, Compile, and Train the Model

```python
#displaying X_train_scaled shape
X_train_scaled.shape[1]
```

```python
# Find the number of columns in the X training data
input_features = X_train_scaled.shape[1]
input_features

# Create the input layer
input = layers.Input(shape=input_features,name="input")

# Create at least two shared layers
shared_layer_1 = layers.Dense(64, activation="relu", name="shared_layer_1")(input)
shared_layer_2 = layers.Dense(128, activation="relu", name="shared_layer_2")(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(32, activation="relu", name="department_hidden")(shared_layer_2)

# Create the output layer
department_output = layers.Dense(3, activation="softmax", name="department_output")(department_hidden)

```

```python
department_output
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(32, activation="relu", name="attrition_hidden")(shared_layer_2)

# Create the output layer
attrition_output = layers.Dense(2, activation="softmax", name="attrition_output")(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer="adam",
              loss={"department_output":"categorical_crossentropy", "attrition_output":"binary_crossentropy"},
              metrics={"department_output":"accuracy", "attrition_output":"accuracy"})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled, {"department_output":y_train_encoded_department, "attrition_output":y_train_encoded_attrition}, epochs=100)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(X_test_scaled, {"department_output":y_test_encoded_department, "attrition_output":y_test_encoded_attrition})
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {results[3]}")
print(f"Attrition Accuracy: {results[4]}")
```

1. Accuracy might not be the best metric to use for this data. The dataset deals with employee attrition and department prediction, where the classes might be imbalanced. In such cases, accuracy can be misleading because it does not take into account the distribution of classes. For instance, if the majority of employees do not leave the company, a model that predicts that no one leaves will have high accuracy but poor performance. Metrics such as precision, recall, F1-score, and AUC-ROC are more informative as they provide a better understanding of the model's performance on imbalanced data by focusing on the true positive and false positive rates.

2. The activation functions chosen for the output layers are:

- Softmax for the Department output layer: Softmax is used because it is suitable for multi-class classification problems where the task is to classify instances into one of three departments. Softmax outputs a probability distribution over the classes, which is appropriate for this scenario.

- Softmax for the Attrition output layer: Although the attrition is a binary classification problem, softmax can still be used to output probabilities for the two classes. However, typically for binary classification, the sigmoid activation function is more common. Softmax can be used here as it ensures the output probabilities sum up to 1, providing a clear probabilistic interpretation for both classes (staying or leaving).
